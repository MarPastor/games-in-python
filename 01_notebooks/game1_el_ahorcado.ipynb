# Juego: El Ahorcado

```python
import random #importamos la libreria random para luego generar las palabras de manera aleatoria



#usamos ASCI para dibujar los distintos estados del ahorcado y que se plasmen visualmente durante el juego
dibujo_ahorcado = ['''
      +---+
      |   |
          |
          |
          |
          |
    ---------''', '''
      +---+
      |   |
      O   |
          |
          |
          |
    ---------''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    ---------''', '''
      +---+
      |   |
      O   |
     -|   |
          |
          |
    ---------''', '''
      +---+
      |   |
      O   |
     -|-  |
          |
          |
    ---------''', '''
      +---+
      |   |
      O   |
     -|-  |
     -    |
          |
    ---------''', '''
      +---+
      |   |
      O   |
     -|-  |
     - -  |
          |
    ---------'''] # Paola: escapar las diagonales

palabra_secreta = ['data', 'gato', 'dato', 'playa', 'algoritmo', 'video'] #definimos las palabras secretas que deberán ser adivinadas
max_errores = 6 #establecemos el máximo de errores que puede cometer el jugador
palabra = random.choice(palabra_secreta).lower() #mediante random seleccionamos aleatoriamente la palabra que deberá ser adivinada
tablero = ['_'] * len(palabra) #establecemos los guiones con la cantidad de letras que tiene la palabra para guiar al jugador
letras_usadas = [] #creamos una lista que almacena las letras que el jugador va ingresando para que no puedan ser repetidas
errores = 0 #creamos un contador de errores que se contrastará con el máximo permitido para permitir jugar o no

"""definimos una función que muestre el dibujo del ahorcado, los guiones que reemplazan las letras de la palabra a adivinar, 
las letras que el jugador ya dijo y el conteo de errores"""

def mostrar_estado_juego(tablero, letras_usadas, errores, max_errores, dibujo_ahorcado): 
    print(dibujo_ahorcado[errores])
    print("Palabra:", " ".join(tablero))
    print("Letras usadas:", ", ".join(letras_usadas))
    print(f"Errores: {errores} de {max_errores}")

"""Definimos una función que va actualizando el tablero a medida que el jugador va adivinando letras. Comienza solo con guiones que 
reemplazan cada letra de la palabra a adivinar y permite visualizar la extensión de la misma, y luego se va completando con las letras
adivinadas en la posición que corresponde dentro de la palabra, hasta reemplazar totalmente los guiones por las letras correctas."""
    
def actualizar_tablero(letra):
    for indice, letra_palabra in enumerate(palabra):
        if letra_palabra == letra:
            tablero[indice] = letra

"""Definimos una función que en cada ronda muestra el estado del ahorcado, las letradas ya usadas y el conteo de errores cometidos y 
permitidos"""

mostrar_estado_juego(tablero, letras_usadas, errores, max_errores, dibujo_ahorcado)

"""Creamos la función central del juego, que nos permite ingresar una letra por ronda, entre A y Z y luego de chequear que no la 
hayamos usado anteriormente, actualiza el tablero posicionándola donde corresponde, muestra el mismo, el estado del dibujo de ahorcado, 
los errores cometidos vs. el total y la adiciona en la lista de letras usadas. 
En caso de que la letra ya la hayamos usado, nos lo informa y solicita otra.
Si la letra no corresponde a la palabra, contabiliza un error y lo suma en la lista de letras usadas, actualizando el tablero.
En caso de se haya alcanzado el límite de 6 errores, informa que perdimos el juego y nos informa cuál era la palabra secreta.
Si por el contrario, la adivinamos, nos felicita."""

while errores <= max_errores and '_' in tablero: #comprueba que el número de errores sea inferior al tope impuesto, y que existan espacios vacíos para adivinar
    
    letra = input("Elige una letra: ").lower()  #pide una letra y la pasa a minúscula para uniformizar

    if ('a' <= letra <= 'z' and len(letra) == 1 and letra in letras_usadas): #chequea que la letra ingresada sea sólo una, entre a y z, y que no esté en la lista de letras usadas
        print("Consejo: debes elegir solo una letra entre a y z que no hayas usado antes.")
        continue  

    elif letra in palabra and letra not in letras_usadas: #si la letra está en la palabra secreta y no está en la lista de letras usadas, actualiza el tablero y el estado del juego
        actualizar_tablero(letra) #actualiza el tablero reemplazando los '_' por la letra en las posiciones correctas
        mostrar_estado_juego(tablero, letras_usadas, errores, max_errores, dibujo_ahorcado) #muestra el estado del dibujo del ahorcado, las letras usadas y los errores cometidos vs. los totales.

    else:
        errores += 1  #si la letra no está en la palabra secreta, suma 1 error
        mostrar_estado_juego(tablero, letras_usadas, errores, max_errores, dibujo_ahorcado) #muestra el dibujo del ahorcado adicionando lo correspondiente al error

    letras_usadas.append(letra)  #guardar la letra ingresada en la lista de letras usadas

    if errores == max_errores: #contrasta que los errores cometidos sean igual al tope establecido
        break
    
if '_' not in tablero: #si no hay más guiones por completar, y no se han cometido la totalidad de errores, entiende que la palabra fue adivinada y le avisa al jugador de la victoria
    print("¡Felicidades, ganaste!")
else:
    print("Perdiste. La palabra era:", palabra) #en caso de los errores cometidos sean superiores a los 6 establecidos como máximo, informa de la derrota

    
```
